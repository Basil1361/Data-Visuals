```python
# Add titles, x, y axis
import matplotlib.pyplot as plt
products = ['A','B','C','D']
sales = ['1000','2000','3000','4000']
colors = ["red", "black", "pink", "purple"]
plt.title("Sales of Product")
plt.ylabel("Price", size = 12)
plt.xlabel("Product", fontsize = 12 , loc = "center")
plt.bar(products, sales, color = colors)
plt.show()
```

```python
# Horizontal bar, just use plt.barh()

import matplotlib.pyplot as plt
products = ['A','B','C','D']
sales = ['1000','2000','3000','4000']
colors = ["red", "black", "pink", "purple"]
plt.title("Sales of Product")
plt.xlabel("Price", size = 12)
plt.ylabel("Product", fontsize = 12 , loc = "center")
# For vertical bars, width is used
# For horizontal bars, width is replaced by height
plt.barh(products, sales, color = colors, height = 0.4)
plt.show()
```

```python
import matplotlib.pyplot as plt

# List of product categories
categories = ["Electronics", "Clothing", "Home Goods", "Books", "Beauty"]

# Sales from Online Store for each category
online_sales = [50000, 35000, 40000, 30000, 25000]

# Sales from Physical Store for each category
physical_sales = [30000, 25000, 35000, 20000, 18000]

w = 0.6

# Add label parameter to each bar
plt.bar(categories, online_sales, width=w, label='Online Sales')
plt.bar(categories, physical_sales, bottom=online_sales, width=w, label='Physical Sales')
plt.xlabel("Product categories")
plt.ylabel("Sales Amount ($)")
plt.title("Comparison of Online and Physical Sales by Category")
plt.legend()  # Now legend will show because we added labels

# ha is horizontal alignment
for i in range(len(categories)):
    plt.text(i, online_sales[i], online_sales[i], ha = "center")
    plt.text(i, physical_sales[i], physical_sales[i], ha = "center")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Electronics', 'Clothing', 'Home Goods', 'Books', 'Beauty']
online_sales = [50000, 35000, 40000, 30000, 25000]
physical_sales = [30000, 25000, 35000, 20000, 18000]

bar_width = 0.6
x = np.arange(len(categories))

fig, ax = plt.subplots(figsize=(8, 4))
bars1 = ax.bar(x, online_sales, bar_width, label='Online Sales', color='#1f77b4')
bars2 = ax.bar(x, physical_sales, bar_width, bottom=online_sales, label='Physical Sales', color='#ff7f0e')

# Add title and labels
ax.set_title('Comparison of Online and Physical Sales by Category', fontsize=14, fontweight='bold')
ax.set_xlabel('Product categories', fontsize=12)
ax.set_ylabel('Sales Amount ($)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=11)

# Add labels on bars
for bar1, bar2 in zip(bars1, bars2):
    # Label for online sales
    ax.text(
        bar1.get_x() + bar1.get_width() / 2,
        bar1.get_height() / 2,
        f'{int(bar1.get_height()):,}',
        ha='center', va='center',
        color='white', fontsize=10, fontweight='bold'
    )
    
    # Label for physical sales
    ax.text(
        bar2.get_x() + bar2.get_width() / 2,
        bar1.get_height() + bar2.get_height() / 2,
        f'{int(bar2.get_height()):,}',
        ha='center', va='center',
        color='black', fontsize=10, fontweight='bold'
    )

# Legend
ax.legend(fontsize=10)

plt.tight_layout()
plt.show()
```
