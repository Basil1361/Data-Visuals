```python
import matplotlib.pyplot as plt
Sport = ["Football","Hockey","Cricket","Basketball","Badminton"]
Students = [10, 15, 10, 5,20]
colorlist = [ 'gold',  'green',  'blue',  'purple',  'orange']

plt.pie(Students, labels = Sport, colors = colorlist, shadow = True, startangle = 140, autopct='%.1f%%')
# percentages with zero decimal places, indicating only whole numbers
plt.title("Favourite Sport of Students", size = 10)
plt.legend(title = "Context", loc = 'upper left')
```
